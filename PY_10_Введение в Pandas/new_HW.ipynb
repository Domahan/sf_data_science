```python
#Создайте признак времени суток поездки time_of_day. Время суток будем определять из часа начала поездки. Условимся, что:
#поездка совершается ночью (night), если её час приходится на интервал от 0 (включительно) до 6 (включительно) часов;
#поездка совершается утром (morning), если её час приходится на интервал от 6 (не включительно) до 12 (включительно) часов;
#поездка совершается днём (day), если её час приходится на интервал от 12 (не включительно) до 18 (включительно) часов;
#поездка совершается вечером (evening), если её час приходится на интервал от 18 (не включительно) до 23 часов (включительно).
#Во сколько раз количество поездок, совершённых днём, больше, чем количество поездок, совёршенных ночью, за представленный в данных период времени?
# Ответ округлите до целых.
def get_time_of_day(time):
    if 0 <= time <= 6:
        return 'night'
    elif 6 < time <= 12:
        return 'morning'
    elif 12 < time <= 18:
        return 'day'
    elif 18 < time <= 23:
        return 'evening'

citibike_tr['time_of_day'] = citibike_tr['starttime'].dt.hour.apply(get_time_of_day)
a = citibike_tr[citibike_tr['time_of_day'] == 'day'].shape[0]
b = citibike_tr[citibike_tr['time_of_day'] == 'night'].shape[0]
print(round(a / b))
```

```python
#Какая риелторская компания (SellerG) имеет наименьшую общую выручку за период с 1 мая по 1 сентября (включительно) 2017 года?
#Для ответа на этот вопрос рассчитайте сумму продаж (Price) каждой компании в заданный период.
# Преобразование столбца Date в datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Задание временного периода
date1 = pd.to_datetime('2017-05-01')
date2 = pd.to_datetime('2017-09-01')

# Правильная фильтрация по дате (включительно)
mask = (melb_df['Date'] >= date1) & (melb_df['Date'] <= date2)

# Группировка по риелтору и расчет суммы продаж, сортировка по возрастанию
revenue_by_seller = melb_df[mask].groupby('SellerG')['Price'].sum().sort_values()

# Вывод компании с наименьшей выручкой
least_revenue_seller = revenue_by_seller.idxmin()
least_revenue_amount = revenue_by_seller.min()
print("\nТоп-5 компаний с наименьшей выручкой:")
print(revenue_by_seller.head(5))
```

```python
#построить таблицу, в которой мы будем учитывать не только число комнат, но и тип здания (Type).
# в параметрах метода groupby() укажем список из нескольких интересующих нас столбцов.
melb_df.groupby(['Rooms', 'Type'])['Price'].mean()
```

```python
# в виде сводной таблицы (первый группировочный признак по строкам, а второй — по столбцам),
#рименяют метод unstack(), который позволяет переопределить вложенный индекс в виде столбцов таблицы:
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

```python
#для построения сводных таблиц существует специальный и более простой метод — pivot_table().
#В качестве параметра values укажем столбец Price, в качестве индексов сводной таблицы возьмём Rooms, а в качестве столбцов — Type. 
# Агрегирующую функцию оставим по умолчанию (среднее). 
# Дополнительно заменим пропуски в таблице на значение 0. 
# Финальный результат для наглядности вывода округлим с помощью метода round() до целых.
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
#построим сводную таблицу, в которой строками будут являться названия регионов (Regionname), 
# а в столбцах будет располагаться наш «признак-мигалка» выходного дня (Weekend), который равен 1, если день был выходным, 
# и 0 — в противном случае. 
# В качестве значений сводной таблицы возьмём количество продаж.
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)
```

```python
#найдём, как зависит средняя и медианная площадь участка (Landsize) от типа объекта (Type) и его региона (Regionname). 
# Чтобы посмотреть несколько статистических параметров, нужно передать в аргумент aggfunc список из агрегирующих функций. 
# Построим такую сводную таблицу, где пропущенные значения заменим на 0:
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

```python
#Давайте построим таблицу, в которой по индексам будут располагаться признаки метода продажи (Method) и типа объекта (Type), 
# по столбцам — наименование региона (Regionname), а на пересечении строк и столбцов будет стоять медианная цена объекта (Price):
melb_df.pivot_table(
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

```python
#Запишем сводную таблицу, которую мы создавали ранее в переменную pivot:
#Выведем её столбцы с помощью атрибута columns:
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
pivot.columns
```

```python
#из таблицы pivot мы можем получить средние значения площадей участков для типа здания unit,
# просто последовательно обратившись по имени столбцов:
display(pivot['mean']['unit'])
```

```python
#Аналогично производится и фильтрация данных. Например, если нам нужны регионы,
# в которых средняя площадь здания для домов типа house меньше их медианной площади, 
# то мы можем найти их следующим образом:
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)
```

```python
#Чтобы получить индексы отфильтрованной таблицы, можно воспользоваться атрибутом index и обернуть результат в список:
print(list(filtered_pivot.index))
```

```python
#Составьте сводную таблицу, которая показывает зависимость медианной цены объекта недвижимости (Price) от риелторского агентства (SellerG) 
# и типа здания (Type).Во вновь созданной таблице найдите агентство, у которого медианная цена для зданий типа unit максимальна. 
# В качестве ответа запишите название этого агентства.
pivot = melb_df.pivot_table(
    values='Price',
    index='SellerG',
    columns='Type',
    aggfunc='median',
    fill_value=0
)
max_unit_price = pivot['unit'].max()
print(pivot[pivot['unit'] == max_unit_price].index[0])
```
