Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

```python
#ваш код здесь
import pandas as pd
def education_level (text):
    first_two_words=" ".join(text.split()[:2]).lower()
    if 'высшее' in first_two_words and 'неоконченное' not in first_two_words:
        return 'высшее'
    elif 'неоконченное высшее' in first_two_words:
        return 'неоконченное высшее'
    elif 'среднее специальное' in first_two_words:
        return 'среднее специальное'
    elif 'среднее' in first_two_words:
        return 'среднее'
    else:
        return 'другое'
    # Создаем новый признак
hh_database['Образование'] = hh_database['Образование и ВУЗ'].apply(education_level)

# Проверяем результат
print(hh_database['Образование'].unique())
print("\nКоличество записей по уровням образования:")
print(hh_database['Образование'].value_counts())

# Удаляем исходный столбец
hh_database.drop('Образование и ВУЗ', axis=1, inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
import pandas as pd
import numpy as np
import re

def experience_to_months(text):
    if pd.isna(text):
        return np.nan
    # Извлекаем только часть до первой запятой (игнорируем периоды работы)
    text = str(text).strip().lower()
    if text == "не указано":
        return np.nan
    main_part = text.split(',')[0]
    # Ищем количество лет
    years = re.search(r'(\d+)\s+год', main_part)
    years = int(years.group(1)) if years else 0
    # Ищем количество месяцев
    months = re.search(r'(\d+)\s+месяц', main_part)
    months = int(months.group(1)) if months else 0
    return years * 12 + months

# Применим функцию
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы'].apply(experience_to_months)
# Проверяем уникальные значения и статистику
print(hh_database['Опыт работы (месяц)'].describe())
print(hh_database['Опыт работы (месяц)'].unique())
# Ответ на контрольный вопрос:
median_months = hh_database['Опыт работы (месяц)'].median()
print(f'Медианный опыт работы (в месяцах) в таблице равен {round(median_months)}')
# Удаляем исходный столбец
hh_database = hh_database.drop(columns=['Опыт работы'])


```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def new_city(arg):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    city = arg.split(' , ')[0]
    if (city == 'Москва') or (city == 'Санкт-Петербург'):
        return city
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
def ready_to_movement(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg):
        return False
    else:
        return True
def ready_for_bisiness_trips(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else: # есть упоминание о командировках и нет отрицания готовности
            return True
    else:  # вообще о командировках нет информации, считаем неготовым
        return False
hh_database['Город'] = hh_database['Город, переезд, командировки'].apply(new_city)
hh_database['Готовность к переезду'] = hh_database['Город, переезд, командировки'].apply(ready_to_movement)
hh_database['Готовность к командировкам'] = hh_database['Город, переезд, командировки'].apply(ready_for_bisiness_trips)
hh_database.drop('Город, переезд, командировки', axis=1, inplace=True)

# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
print('В Санкт-Петербурге живет', \
      round(hh_database['Город'].value_counts(normalize=True)['Санкт-Петербург'] * 100),'процентов соискателей')
#Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
both_ready = hh_database[
    (hh_database['Готовность к переезду'] == True) & 
    (hh_database['Готовность к командировкам'] == True)
].shape[0]

# Расчет процента
percentage = (both_ready / len(hh_database)) * 100

print(f'Готовы одновременно и к переездам, и к командировкам: {round(percentage)}%')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# Определяем все возможные категории
employment_categories = ['полная занятость', 'частичная занятость','проектная работа', 'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график','удаленная работа', 'вахтовый метод']
# Функция для создания признаков-мигалок
def one_hot_encode(df, column, categories):
    for category in categories:
        df[f'{column}_{category}'] = df[column].str.contains(category, na=False)
    return df
# Применяем One Hot Encoding
hh_database = one_hot_encode(hh_database, 'Занятость', employment_categories)
hh_database = one_hot_encode(hh_database, 'График', schedule_categories)
# Сколько людей ищут проектную работу и волонтерство (в обоих столбцах стоит True)?
project_and_volunteer = hh_database[
    hh_database['Занятость_проектная работа'] & hh_database['Занятость_волонтерство']].shape[0]

print(f"Количество соискателей, ищущих и проектную работу, и волонтерство: {project_and_volunteer}")
#Сколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)?
shift_and_flexible = hh_database[
    hh_database['График_вахтовый метод'] & hh_database['График_гибкий график']].shape[0]

print(f"Количество соискателей, желающих работать вахтовым методом и с гибким графиком: {shift_and_flexible}")
# Удаляем исходные столбцы
hh_database.drop(['Занятость', 'График'], axis=1, inplace=True)
```

```python
# ваш код здесь
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

def experience_to_months(text):
    if pd.isna(text):
        return np.nan
    # Извлекаем только часть до первой запятой (игнорируем периоды работы)
    text = str(text).strip().lower()
    if text == "не указано":
        return np.nan
    main_part = text.split(',')[0]
    # Ищем количество лет
    years = re.search(r'(\d+)\s+год', main_part)
    years = int(years.group(1)) if years else 0
    # Ищем количество месяцев
    months = re.search(r'(\d+)\s+месяц', main_part)
    months = int(months.group(1)) if months else 0
    return years * 12 + months

# Применим функцию
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы'].apply(experience_to_months)
# Проверяем уникальные значения и статистику
print(hh_database['Опыт работы (месяц)'].describe())


# Устанавливаем разумный предел (50 лет = 600 месяцев)
max_reasonable = 50 * 12
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы (месяц)'].apply(
    lambda x: min(x, max_reasonable) if pd.notna(x) else x
)
print('Мода опыта работы:',hh_database['Опыт работы (месяц)'].mode()[0])
# Создаем subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Гистограмма
ax1.hist(hh_database['Опыт работы (месяц)'].dropna(), bins=20, color='skyblue', 
         alpha=0.7, edgecolor='black')
ax1.set_title('Распределение Опыт работы (месяц)')
ax1.set_xlabel('Опыт работы (месяц)')
ax1.set_ylabel('Частота')
ax1.grid(True, alpha=0.3)

# Boxplot
sns.boxplot(data=hh_database, y='Опыт работы (месяц)', ax=ax2, orient='v')
ax2.set_title('Boxplot Опыт работы (месяц)')
ax2.set_ylabel('Опыт работы (месяц)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
# Удаляем исходный столбец
hh_database = hh_database.drop(columns=['Опыт работы'])
```

```python
# ваш код здесь

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def get_salary_num(arg):
    if pd.isna(arg):
        return np.nan
    try:
        salary = float(arg.split(' ')[0])
        return salary
    except:
        return np.nan

def get_salary_currency(arg):
    if pd.isna(arg):
        return np.nan
    
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN',
        'руб': 'RUB'
    }
    
    try:
        curr = arg.split(' ')[1].replace('.', '')
        return currency_dict.get(curr, np.nan)
    except:
        return np.nan

# Загрузка и обработка данных
rates = pd.read_csv('data/ExchangeRates.csv')
rates['date'] = pd.to_datetime(rates['date'], dayfirst=True).dt.date
hh_database['Обновление резюме'] = pd.to_datetime(hh_database['Обновление резюме'], dayfirst=True).dt.date

hh_database['ЗП (tmp)'] = hh_database['ЗП'].apply(get_salary_num)
hh_database['Курс (tmp)'] = hh_database['ЗП'].apply(get_salary_currency)

merged = pd.merge(
    hh_database, rates, how='left',
    left_on=['Курс (tmp)', 'Обновление резюме'],
    right_on=['currency', 'date']
)

merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
hh_database['ЗП (руб)'] = merged['close'] * merged['ЗП (tmp)'] / merged['proportion']

data = hh_database.drop(['ЗП', 'ЗП (tmp)', 'Курс (tmp)'], axis=1)

# Анализ распределения ЗП
salary_data = hh_database['ЗП (руб)'].dropna()

# Определяем границы выбросов
Q1 = salary_data.quantile(0.25)
Q3 = salary_data.quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Фильтруем данные без выбросов
filtered_salary = salary_data[salary_data <= upper_bound]

print("АНАЛИЗ РАСПРЕДЕЛЕНИЯ ЗАРПЛАТЫ:")
print(f"Медианная зарплата: {round(salary_data.median()/1000)} тыс. руб")
print(f"Минимальная зарплата: {salary_data.min():,.0f} руб")
print(f"Максимальная зарплата: {salary_data.max():,.0f} руб")
print(f"Мода зарплата: {salary_data.mode()[0]:,.0f} руб")
print(f"Количество наблюдений: {len(salary_data)}")
print(f"Верхняя граница без выбросов: {upper_bound:,.0f} руб")
print(f"Количество выбросов: {len(salary_data) - len(filtered_salary)}")

# Построение графиков БЕЗ ВЫБРОСОВ
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Гистограмма без выбросов
ax1.hist(filtered_salary, bins=30, color='skyblue', alpha=0.7, edgecolor='black')
ax1.set_title('Распределение желаемой заработной платы (без выбросов)')
ax1.set_xlabel('ЗП (руб)')
ax1.set_ylabel('Количество соискателей')
ax1.grid(True, alpha=0.3)

# Boxplot без выбросов
sns.boxplot(y=filtered_salary, ax=ax2)
ax2.set_title('Boxplot заработной платы (без выбросов)')
ax2.set_ylabel('ЗП (руб)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
# ваш код здесь
# Сохраняем исходный размер данных
initial_shape = hh_database_cleaned.shape
print(f"Исходный размер данных: {initial_shape}")

def get_salary_num(arg):
    salary = float(arg.split(' ')[0])
    return salary

def get_salary_currency(arg):
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    curr = arg.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency_dict[curr]
    
rates = pd.read_csv('data/ExchangeRates.csv')
rates['date'] = pd.to_datetime(rates['date'], dayfirst=True).dt.date
hh_database_cleaned['Обновление резюме'] = pd.to_datetime(hh_database_cleaned['Обновление резюме'], dayfirst=True).dt.date
hh_database_cleaned['ЗП (tmp)'] = hh_database_cleaned['ЗП'].apply(get_salary_num)
hh_database_cleaned['Курс (tmp)'] = hh_database_cleaned['ЗП'].apply(get_salary_currency)
merged = pd.merge(
    hh_database_cleaned, rates, how='left',
    left_on=['Курс (tmp)', 'Обновление резюме'],
    right_on=['currency', 'date',], suffixes=('', '_rates')
)
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
hh_database_cleaned['ЗП (руб)'] = merged['close'] * merged['ЗП (tmp)'] / merged['proportion']
data = hh_database_cleaned.drop(['ЗП', 'ЗП (tmp)', 'Курс (tmp)'], axis=1)

# Анализ выбросов до очистки 
too_low_before = (hh_database_cleaned['ЗП (руб)'] < 1000).sum()
too_high_before = (hh_database_cleaned['ЗП (руб)'] > 1000000).sum()
print(f"Записей с зарплатой < 1,000 руб.: {too_low_before}")
print(f"Записей с зарплатой > 1,000,000 руб.: {too_high_before}")
    
# Удаляем резюме с аномальной зарплатой
salary_filter = (hh_database_cleaned['ЗП (руб)'] >= 1000) & (hh_database_cleaned['ЗП (руб)'] <= 1000000)
before_cleaning = hh_database_cleaned.shape[0]  # Сохраняем размер до очистки
hh_database_cleaned = hh_database_cleaned[salary_filter]
after_cleaning = hh_database_cleaned.shape[0]   # Сохраняем размер после очистки
# Выводим результаты очистки
removed_count = before_cleaning - after_cleaning  # Правильно вычисляем количество удаленных    
# Выводим результаты очистки
final_shape = hh_database_cleaned.shape
print(f"\nРЕЗУЛЬТАТЫ ОЧИСТКИ ОТ ВЫБРОСОВ:")
print("=" * 50)
print(f"Удалено резюме с аномальной зарплатой: {removed_count}")
print(f"Новый размер данных: {final_shape}")
```

5. В процессе разведывательного анализа мы обнаружили резюме, в которых **опыт работы в годах превышал возраст соискателя**. Найдите такие резюме и удалите их из данных


```python
# ваш код здесь
# Находим резюме, где опыт работы превышает возраст соискателя
print("ПОИСК АНОМАЛЬНЫХ РЕЗЮМЕ:")
print("=" * 40)

# Создаем функцию для извлечения возраста из столбца 'Пол, возраст'
def extract_age(gender_age_str):
    try:
        # Ищем паттерн "число лет"
        age_match = re.search(r'(\d+)\s+лет', str(gender_age_str))
        if age_match:
            return int(age_match.group(1))
        return None
    except:
        return None

# Создаем функцию для извлечения опыта работы из столбца 'Опыт работы'
def extract_experience_years(experience_str):
    try:
        # Ищем годы опыта работы
        years_match = re.search(r'Опыт работы\s+(\d+)\s+лет', str(experience_str))
        months_match = re.search(r'(\d+)\s+месяц', str(experience_str))
        
        years = int(years_match.group(1)) if years_match else 0
        months = int(months_match.group(1)) if months_match else 0
        
        return years + months/12  # возвращаем в годах с дробной частью
    except:
        return None

# Добавляем новые столбцы
hh_database['Возраст_число'] = hh_database['Пол, возраст'].apply(extract_age)
hh_database['Опыт_лет'] = hh_database['Опыт работы'].apply(extract_experience_years)

# Находим аномалии: опыт работы > возраст
anomaly_condition = hh_database['Опыт_лет'] > hh_database['Возраст_число']
anomalies = hh_database[anomaly_condition]

print(f"Найдено аномальных резюме: {len(anomalies)}")

if len(anomalies) > 0:
    print("\nАНАМАЛЬНЫЕ РЕЗЮМЕ:")
    print("=" * 30)
    
    # Показываем аномальные записи
    for idx, row in anomalies.iterrows():
        print(f"Резюме #{idx}:")
        print(f"  Возраст: {row['Возраст_число']} лет")
        print(f"  Опыт работы: {row['Опыт_лет']:.1f} лет")
        print(f"  Должность: {row['Ищет работу на должность:']}")
        print(f"  Зарплата: {row['ЗП']}")
        print("-" * 50)
    
    # Удаляем аномальные записи
    initial_count = len(hh_database)
    hh_database_cleaned = hh_database[~anomaly_condition]
    final_count = len(hh_database_cleaned)
    
    print(f"\nРЕЗУЛЬТАТЫ ОЧИСТКИ:")
    print("=" * 30)
    print(f"Удалено аномальных резюме: {initial_count - final_count}")
    print(f"Осталось резюме: {final_count}")
    print(f"Процент удаленных: {(initial_count - final_count)/initial_count*100:.2f}%")
    
    # Обновляем основной DataFrame
    hh_database = hh_database_cleaned.copy()
    
else:
    print("Аномальных резюме не найдено!")
```
